# Merge csv and json data

```python
import json
import pandas as pd
import numpy as np
```

```python
# Opening JSON file
f = open('./data/zalo/test/public_test.json')
# returns JSON object as 
# a dictionary
data = json.load(f)
# Closing file
f.close()
```

```python
data_1 = pd.read_csv("./data/zalo/test/zac2023-math-public-test - zac2023-math-public-test.csv")
data_1 = data_1.fillna("None")
```

```python
data_1
```

```python
def convert_string_to_list(text: str) -> list:
    result = []
    start = None
    end = None
    for i, char in enumerate(text):
        if char=="'":
            if start is None:
                start = i+1
            else:
                end = i
                result.append(text[start:end])
                start = None
                end = None
    
    return result

data_1["choices"] = data_1["choices"].apply(convert_string_to_list)
```

```python
ids = []
questions = []
choices_list = []
answers = []
for i, id, ques, choices, answer, _ in data_1.itertuples():
    ids.append(id)
    questions.append(ques)
    choices_list.append(choices)
    if answer=="A" or answer=="?":
        answers.append(choices[0])
    elif answer=="B":
        answers.append(choices[1])
    elif answer=="C":
        answers.append(choices[2])
    elif answer=="D":
        answers.append(choices[3])
    else:
        answers.append(answer)
data_1 = pd.DataFrame({"id": ids, "question": questions, "choices": choices_list, "answer": answers})
```
